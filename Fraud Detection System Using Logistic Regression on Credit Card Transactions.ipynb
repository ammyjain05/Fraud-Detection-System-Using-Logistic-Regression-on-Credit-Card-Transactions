#Importing the libraires 

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

#Loading the csv file

```python
cc_data = pd.read_csv('creditcard.csv')
```

```python
cc_data.head()
```

```python
cc_data.tail()
```

```python
cc_data.info()
```

```python
#checking missing values in each column
cc_data.isnull().sum()
```

```python
#distrubution of legit and fraudulent transactions
# 0 -> legit
# 1 -> Fraud

cc_data['Class'].value_counts()
```

#Data Seperation

```python
legit = cc_data[cc_data.Class == 0]
fraud = cc_data[cc_data.Class == 1]
```

```python
legit.count().value_counts()
```

```python
print(fraud.shape)
```

```python
#Statistical measures in dollars
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#Compare the values
cc_data.groupby('Class').mean()
```

```python
#Building new dataset as the distrubiton 
#of transaction values is very large
#sampling value randomly

l_sample = legit.sample(n=492)
```

```python
#joining the data frame
#axis = 0 - row wise, axis = 1 - column.

final_data = pd.concat([l_sample,fraud], axis=0)
```

```python
final_data.head()
```

```python
final_data['Class'].value_counts()
```

```python
#to check if we got a good/bad sample
#the mean for the origianl data and here is almost
#similar which tells us the natures of the data is same
#therefore our ML model will give a good result

final_data.groupby('Class').mean()
```

split the data

```python
X = final_data.drop(columns = 'Class', axis=1)
Y = final_data['Class']
```

```python
print(X)
```

```python
print(Y)
```

#Train and Test

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=3)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

#Model Training
#Logistic Regression
#Training the Data

```python
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

#Evaluation of the model

```python
X_train_prediction = model.predict(X_train)
train_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print("Training Accuracy: ", train_accuracy)
```

```python
X_test_prediction = model.predict(X_test)
test_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print("Training Accuracy: ", test_accuracy)
```
